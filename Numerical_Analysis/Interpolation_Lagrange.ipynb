```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.cos((np.pi - x) / 2) - x
```

## Lagrange's polynomial

$\Large
L_n(x) = \sum_{i=0}^n f(x_i) \frac{ (x - x_0)...(x - x_{k-1})(x - x_{k+1})...(x - x_n) }{ (x_k - x_0)...(x_k - x_{k-1})(x_k - x_{k+1})...(x_k - x_n)}
$

```python
n = 10
X = np.linspace(a, b, n) # узлы интерполяции (interpolation nodes)

def Ln(x):
    sum = np.zeros(x.shape[0])
    
    for i in range(n):
        prod = f(X[i])
        for j in range(n):
            if i != j:
                prod *= (x - X[i]) / (X[i] - X[j])
        sum += prod
        
    return sum
```
